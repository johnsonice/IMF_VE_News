```python
def read_grouped_search_words(file_path):
    """
    Read the search words, by group, from the indicated file
    :param file_path:
    :return: 'dict' mapping key(group name) -> value(list of str words-in-group)
    """

    df = pd.read_csv(file_path)
    search_groups = df.to_dict()
    for k, v in search_groups.items():
        temp_list = [i for i in list(v.values()) if not pd.isna(i)]  # Save all non-NA values - different len groups
        temp_list = [wg.split('&') for wg in temp_list]   # split & for wv search
        search_groups[k] = temp_list  # Map str key group name -> list[str] group words
    return search_groups

grouped_search_words_file = "C:\\Users\\Alex P\\Desktop\\IMF\\research_bkp\\research\\test\\group_words_using_w2v_on_sentiment.csv"
```

```python
import copy
sim_word_map = {}
# Read base sentiment definitions
sim_wrds_df = pd.read_csv(grouped_search_words_file).drop(columns=['Unnamed: 0', 'recall', 'prec', 'f2']).set_index('word')
for wrd in sim_wrds_df.index:
    splitted = sim_wrds_df.loc[wrd].sim_words.split(',')
    trimmed = []
    for spl in splitted:
        stripped = spl.strip()
        trmd = ''
        for ch in stripped:
            if ch in ["'", '\"', '[', ']']:
                continue
            trmd += ch
        trimmed.append(trmd)
    sim_word_map[wrd] = trimmed
sim_word_map
```

```python
# Aggregate agg definitions
pos_sent = 'positive_sentiment_language'
neg_sent = 'negative_sentiment_language'
base_sents = list(sim_word_map.keys())

base_sents.remove(pos_sent)
base_sents.remove(neg_sent)
agg_all_other = []
for sent in base_sents:
    agg_all_other.extend(sim_word_map[sent])
sim_word_map['agg_all_other_sentiments'] = agg_all_other
agg_other_and_negative = copy.deepcopy(agg_all_other)
agg_other_and_negative.extend(sim_word_map['negative_sentiment_language'])
sim_word_map['agg_other_and_negative'] = agg_other_and_negative
all_language = copy.deepcopy(agg_other_and_negative)
all_language.extend(sim_word_map)
sim_word_map['all_language'] = all_language

sim_word_map
#eval_df = evaluate()
```

```python
def sum_month(month_df, to_sum):
    tally = 0
    for wrd in to_sum:
        if wrd in month_df.index:
            tally += month_df[wrd]
    return tally

def create_ts_df(ctry, freq_df, ts_defs):
    new_cols = list(ts_defs.keys())
    months = freq_df['month'].values
    freq_df_mi = freq_df.set_index('month')
    data = np.ndarray((len(months),len(ts_defs)), np.double)
    for j, col in enumerate(new_cols):
        to_sum = ts_defs[col]
        for i, mnth in enumerate(months):
            month_df = freq_df_mi.loc[mnth]
            summed = sum_month(month_df, to_sum)
            data[i,j] = summed
    #to_re = pd.DataFrame(index=months, columns=ts_defs, data=data).fillna(value=0)
    to_re = pd.DataFrame(index=months, columns=ts_defs, data=data)
    
    return to_re
    
def create_eval_df(ctry, ts_df, crisis_def='kr', eval_end_date=None):
    if not eval_end_date:
        eval_end_date = {'q': '2001Q4',
                     'm': '2001-12'} # KR end date (original)
    
    data = []
    for sent in ts_df.columns:
        sent_tf = ts_df[sent]
        recall, precision, fscore, \
        _, _, _ = evaluate(frequency_ser=sent_tf, country=ctry, method='zscore', 
                                 crisis_defs=crisis_def, period='month',stemmed=False, 
                                 window=24, direction='incr', months_prior=24,
                                 fbeta=2,eval_end_date=eval_end_date, weights=None,
                                 z_thresh=2.1)
        data.append([recall, precision, fscore]) #TODO pickup
    to_re = pd.DataFrame(index=ts_df.columns, columns=['Recall', 'Precision', 'F2 Score'],
                        data=data)
    return to_re
```

```python
def get_crises(crisis_def, sub_select = 'all'):
    # Return the countries associated with each crisis definition
    if crisis_def == 'kr':
        if sub_select == 'all':
            return crisis_points.crisis_points_TEMP_KnR
        elif sub_select == 'j5':
            all_ctry_crisis = crisis_points.crisis_points_TEMP_KnR
            j5_ctry = list(all_ctry_crisis.keys())[:5]
            j5_crises = {}
            for j5c in j5_ctry:
                j5_crises[j5c] = all_ctry_crisis[j5c]
            return j5_crises
    if crisis_def == 'll':
        return crisis_points.ll_crisis_points
    if crisis_def == 'IMF_GAP_6':
        return crisis_points.imf_gap_6_events
    if crisis_def == 'IMF_GAP_0':
        return crisis_points.imf_all_events
    if crisis_def == 'RomerRomer':
        return crisis_points.crisis_points_RomerNRomer
    if crisis_def == 'LoDuca':
        return crisis_points.crisis_points_LoDuca
    if crisis_def == 'ReinhartRogoff':
        return crisis_points.crisis_points_Reinhart_Rogoff_All
    if crisis_def == 'IMF_Monthly_Starts':
        return crisis_points.imf_programs_monthly
    if crisis_def == 'IMF_Monthly_Starts_Gap_3':
        return crisis_points.imf_programs_monthly_gap3
    if crisis_def == 'IMF_Monthly_Starts_Gap_6':
        return crisis_points.imf_programs_monthly_gap6

def app_num_crises(ctry, crises):
    num_crises = len(crises[ctry]['starts'])
    return num_crises

def add_num_crises_in_ctry(to_add_df, crisis_def, sub_select='all'):
    crises = get_crises(crisis_def, sub_select)
    cp_df = to_add_df.copy()
    cp_df['Num Crises'] = cp_df['Country'].apply(app_num_crises, args=[crises])
    return cp_df
```

```python
# Apply Weights
mthd_ctry_df_fill_weight_app = mthd_ctry_df_fill_weight.copy()
total_crises = mthd_ctry_df_fill_weight.groupby(['Identification Scheme', 'Sentiment']).sum()['Num Crises'].values[0]
mthd_ctry_df_fill_weight_app['Recall'] = mthd_ctry_df_fill_weight_app['Recall']* mthd_ctry_df_fill_weight_app['Num Crises']/total_crises
mthd_ctry_df_fill_weight_app['Precision'] = mthd_ctry_df_fill_weight_app['Precision']* mthd_ctry_df_fill_weight_app['Num Crises']/total_crises
mthd_ctry_df_fill_weight_app['F2 Score'] = mthd_ctry_df_fill_weight_app['F2 Score']* mthd_ctry_df_fill_weight_app['Num Crises']/total_crises
# Average with weights
weighted_cross_country_avg = mthd_ctry_df_fill_weight_app[['Identification Scheme', 'Recall', 'Precision', 'F2 Score']].groupby(['Identification Scheme']).mean()
weighted_cross_country_avg['Country'] = 'Weighted Cross-Country Mean'
weighted_cross_country_avg
```

```python
def create_rand_ts_df(ctry, freq_df, rand_length=rand_len):
    new_cols = ['Random Benchmark']
    months = freq_df['month'].values
    freq_df_mi = freq_df.set_index('month')
    data = np.ndarray((len(months),len(ts_defs)), np.double)
    for j, col in enumerate(new_cols):
        to_sum = ts_defs[col]
        for i, mnth in enumerate(months):
            month_df = freq_df_mi.loc[mnth]
            summed = sum_month(month_df, to_sum)
            data[i,j] = summed
    #to_re = pd.DataFrame(index=months, columns=ts_defs, data=data).fillna(value=0)
    to_re = pd.DataFrame(index=months, columns=ts_defs, data=data)
    
    return to_re
    
def create_eval_df(ctry, ts_df, crisis_def='kr', eval_end_date=None):
    if not eval_end_date:
        eval_end_date = {'q': '2001Q4',
                     'm': '2001-12'} # KR end date (original)
    
    data = []
    for sent in ts_df.columns:
        sent_tf = ts_df[sent]
        recall, precision, fscore, \
        _, _, _ = evaluate(frequency_ser=sent_tf, country=ctry, method='zscore', 
                                 crisis_defs=crisis_def, period='month',stemmed=False, 
                                 window=24, direction='incr', months_prior=24,
                                 fbeta=2,eval_end_date=eval_end_date, weights=None,
                                 z_thresh=2.1)
        data.append([recall, precision, fscore]) #TODO pickup
    to_re = pd.DataFrame(index=ts_df.columns, columns=['Recall', 'Precision', 'F2 Score'],
                        data=data)
    return to_re
```
