```python
# Do before multi-indexing
def get_word_count_def(sent_x_def):
     return sent_x_def.split('x_')[1]
# Add base word counts def (fed, w2v, w2v_refined_2 - positives and negatives)
def add_word_count_def(df):
    # base = None
    df['word_count_def'] = None
    
    # first six == sentiment
    df.loc[0:6,['word_count_def']] = df.loc[0:6]['sentiment']

    # next nine are none
    df.loc[6:15, ['word_count_def']] = None
    
    # Rest are [sentiment_def]_x_[word_count_def]
    df.loc[15:, ['word_count_def']] = df.loc[15:]['sentiment'].apply(get_word_count_def)

def get_count_base(count_def):
    if count_def is None:
        return None
    splitted = count_def.split('_')
    return '_'.join(splitted[:-1])
# Add word count bases (fed, w2v, w2v_refined_2)
def add_word_count_base(df):
    # base = None
    df['word_count_base'] = df['word_count_def'].apply(get_count_base)

def get_sentiment_kind(sent_x_def):
    before_count = len(sent_x_def.split('_x')[0].split('_'))
    if before_count == 2:
        return 'intensity'
    if before_count == 3:
        return 'direction'
    
# Add sentiment kind vader_pos_x_fed_pos 
def add_sentiment_kind(df):
    # base = None
    df['sentiment_kind'] = None
    
    # First six (0-5) is just word counting
    df.loc[0:6,['sentiment_kind']] = 'word-count'
    
    # Next nine are special
    # 6-8 intensity
    df.loc[6:9,['sentiment_kind']] = 'intenstiy'
    # 9 direction
    df.loc[9,['sentiment_kind']] = 'direction'
    # 10-11 intensity
    df.loc[10:12,['sentiment_kind']] = 'intenstiy'
    # 13 number of document
    df.loc[13,['sentiment_kind']] = 'document-count'
    # 14 direction
    df.loc[14,['sentiment_kind']] = 'direction'
    
    # Rest are [sentiment_def]_x_[word_count_def]
    # Generated in a format - 3 words before x => direction, 2 words => intensity   
    df.loc[15:, ['sentiment_kind']] = df.loc[15:]['sentiment'].apply(get_sentiment_kind)

def add_direction(df):
    # 8, 10, 11, 13 are neither
    #exceptions = [9, 10, 11, 13]
    #df.loc[exceptions, ['is_positive']] = None
    df['is_positive'] = None
    
    # Even, except for 8-13 are positive, up to 14
    # also 9, 12
    even = list(range(0, 15, 2))
    df.loc[even, 'is_positive'] = True
    
    # Odd are negative, except 8, 10, 11, 13, up to 14
    odds = list(range(1, 14, 2))
    df.loc[odds,['is_positive']] = False
    
    # From 15:, odds positive, evens negative
    l_odd = list(range(15, len(df), 2))
    l_even = list(range(16, len(df), 2))
    df.loc[l_even,['is_positive']] = False
    df.loc[l_odd,['is_positive']] = True  
    
# Add is_compound_sentiment
def add_is_compound_sentiment(df):
    df['is_compound_sentiment'] = False
    df.loc[15:, ['is_compound_sentiment']] = True

# Def add_improvement_over_base
def add_improvement_over_base(df):
    df['improvement_recall'] = 0
    df['improvement_precision'] = 0
    df['improvement_fscore'] = 0
    df['improvement_recall_%'] = 0
    df['improvement_precision_%'] = 0
    df['improvement_fscore_%'] = 0
    
    word_count_defs = df['word_count_def'].dropna().unique()
    for definition in word_count_defs:
        base_compare = df[(df['word_count_def']==definition) & (df['is_compound_sentiment']==False)]
        to_compare =  df[(df['word_count_def']==definition) & (df['is_compound_sentiment']==True)]
        for metric in ['recall','precision','fscore']:
            metric_base = base_compare[metric].values[0]
            improvement = to_compare[metric].values - metric_base
            improve_indeces = to_compare.index
            #print('BASE',metric_base)
            #print('COMPARE',to_compare[metric])
            #print(improvement)
            df.loc[improve_indeces, [f'improvement_{metric}']] = improvement
            df.loc[improve_indeces, [f'improvement_{metric}_%']] = improvement/metric_base * 100

# Add sentimenter_name
def get_sentimenter_name(sent_x_def):
     return sent_x_def.split('_')[0]
    
# Add Sentimenter name
def add_sentimenter_name(df):
    # base = None
    df['sentimenter_name'] = None
    
    # Rest are [sentiment_def]_x_[word_count_def]
    df.loc[15:, ['sentimenter_name']] = df.loc[15:]['sentiment'].apply(get_sentimenter_name)


# Create pairs df
def pair_improvements(df, in_percent=True):
    if in_percent:
        df_out = df.groupby(['crisis_definition','word_count_base', 'sentimenter_name', 'sentiment_kind'])['improvement_recall_%','improvement_precision_%','improvement_fscore_%'].mean()        
    else:
        df_out = df.groupby(['crisis_definition', 'word_count_base', 'sentimenter_name', 'sentiment_kind'])['improvement_recall','improvement_precision','improvement_fscore'].mean()    
    return df_out
```

```python
def prep_df(df):
    add_word_count_def(df)
    add_word_count_base(df)
    add_sentimenter_name(df)
    add_sentiment_kind(df)
    add_word_count_base(df)
    add_direction(df)
    add_is_compound_sentiment(df)
    add_improvement_over_base(df)
```

```python
def mean_corr(list_of_corr_df):
    mega_df = pd.concat(list_of_corr_df)
    mega_df = mega_df.groupby('Unnamed: 0', sort=False).mean().reset_index()
    return mega_df

corr_df = mean_corr(corr_dfs)
corr_df
```

```python
def add_correlations_paired_improvement(df, corr_df):
    
    df['correlation'] = np.NaN
    df = df.reset_index()

    # Fed Words
    df.loc[[0],['correlation']] = corr_df.iloc[9]['Correlation']
    df.loc[[1],['correlation']] = corr_df.iloc[8]['Correlation']
    df.loc[[2],['correlation']] = corr_df.iloc[7]['Correlation']
    df.loc[[3],['correlation']] = corr_df.iloc[6]['Correlation']
    df.loc[[4],['correlation']] = corr_df.iloc[5]['Correlation']
    df.loc[[5],['correlation']] = corr_df.iloc[4]['Correlation']

    # w2v words
    df.loc[[0+6],['correlation']] = corr_df.iloc[9+6]['Correlation']
    df.loc[[1+6],['correlation']] = corr_df.iloc[8+6]['Correlation']
    df.loc[[2+6],['correlation']] = corr_df.iloc[7+6]['Correlation']
    df.loc[[3+6],['correlation']] = corr_df.iloc[6+6]['Correlation']
    df.loc[[4+6],['correlation']] = corr_df.iloc[5+6]['Correlation']
    df.loc[[5+6],['correlation']] = corr_df.iloc[4+6]['Correlation']
    
    # Revised w2v words
    df.loc[[0+12],['correlation']] = corr_df.iloc[9+12]['Correlation']
    df.loc[[1+12],['correlation']] = corr_df.iloc[8+12]['Correlation']
    df.loc[[2+12],['correlation']] = corr_df.iloc[7+12]['Correlation']
    df.loc[[3+12],['correlation']] = corr_df.iloc[6+12]['Correlation']
    df.loc[[4+12],['correlation']] = corr_df.iloc[5+12]['Correlation']
    df.loc[[5+12],['correlation']] = corr_df.iloc[4+12]['Correlation']
    
    return df.set_index(['crisis_definition', 'word_count_base', 'sentimenter_name', 'sentiment_kind'])
    
add_correlations_paired_improvement(pair_test, corr_df)
```

```python
def load_crises(df_list, corr_df):
    done_dfs = []
    for df in df_list:
        prep_df(df)
        df = pair_improvements(df)
        df = add_correlations_paired_improvement(df, corr_df)
        done_dfs.append(df)
    out_df = pd.concat(done_dfs)
    return out_df
```
